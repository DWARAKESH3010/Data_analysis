```python
import cv2
import numpy as np
cap = cv2.VideoCapture(r"D:\Video\video3.mp4")

fps = cap.get(cv2.CAP_PROP_FPS)
delay = int(500 / fps)
print(fps)
while(1):
    _, frame = cap.read()
    # It converts the BGR color space of image to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Threshold of blue in HSV space
    lower_blue = np.array([60, 35, 140])
    upper_blue = np.array([180, 255, 255])

    # preparing the mask to overlay
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    
    # The black region in the mask has the value of 0,
    # so when multiplied with original image removes all non-blue regions
    result = cv2.bitwise_and(frame, frame, mask = mask)

    cv2.imshow('frame', frame)
    cv2.imshow('mask', mask)
    cv2.imshow('result', result)

    cv2.waitKey(10)
    

cv2.destroyAllWindows()
cap.release()
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    img = frame
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 110, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Loop through contours
    for contour in contours:
        # Draw the contour
        cv2.drawContours(img, [contour], -1, (0, 255, 0), 2)

        # Loop through each point in the contour and draw a small circle + coordinate
        for point in contour:
            x, y = point[0]
            cv2.circle(img, (x, y), 3, (255, 0, 0), -1)
            cv2.putText(img, f"{x},{y}", (x+5, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1)

    # Show the result
    cv2.imshow('frame',img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# Clean-up squad
cap.release()
cv2.destroyAllWindows()



```

```python
import cv2
import numpy as np

# Open webcam (0 = default camera)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Read frame-by-frame
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply blur to reduce noise
    gray_blurred = cv2.blur(gray, (3, 3))

    # Apply Hough Circle Transform
    circles = cv2.HoughCircles(gray_blurred, 
                               cv2.HOUGH_GRADIENT, 
                               dp=1, minDist=20, 
                               param1=50, param2=30, 
                               minRadius=1, maxRadius=40)

    # If circles are found
    if circles is not None:
        circles = np.uint16(np.around(circles))

        for pt in circles[0, :]:
            a, b, r = pt[0], pt[1], pt[2]

            # Draw the circle's outer boundary
            cv2.circle(frame, (a, b), r, (0, 255, 0), 2)

            # Draw the center of the circle
            cv2.circle(frame, (a, b), 1, (0, 0, 255), 3)

    # Show the frame
    cv2.imshow("Live Circle Detection", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Start the webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect corners
    corners = cv2.goodFeaturesToTrack(gray, maxCorners=27, qualityLevel=0.01, minDistance=10)
    
    if corners is not None:
        corners = np.int0(corners)

        for i in corners:
            x, y = i.ravel()
            cv2.circle(frame, (x, y), 4, (0, 255, 0), -1)

    # Show the frame with corners
    cv2.imshow('Corners in Live Video', frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the noisy image
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\Saved Pictures\noise_image.jpg")

median = cv2.medianBlur(img, 3)
bilateral = cv2.bilateralFilter(median, d = 9, sigmaColor = 100, sigmaSpace = 100)
final = cv2.fastNlMeansDenoisingColored(bilateral, None, 10, 10, 7, 21)

# Sharpen the denoised image
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
final_sharp = cv2.filter2D(final, -1, kernel)

# Plot
plt.figure(figsize=(18,8))
plt.subplot(1, 3, 1)
plt.title("Noisy Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Denoised Image")
plt.imshow(cv2.cvtColor(final, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Denoised + Sharpened")
plt.imshow(cv2.cvtColor(final_sharp, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```
