```python
# blurring the image
cv2.imshow('Original Image', read)
cv2.waitKey(0)

# Gaussian Blur
Gaussian = cv2.GaussianBlur(read, (7, 7), 0)
cv2.imshow('Gaussian Blurring', Gaussian)
cv2.waitKey(0)

# Median Blur
median = cv2.medianBlur(read, 5)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)


# Bilateral Blur
bilateral = cv2.bilateralFilter(read, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
# border of the image
image = cv2.copyMakeBorder(read, 10, 10, 10, 10, cv2.BORDER_CONSTANT, None, value=(0, 255, 0)) #green
cv2.imshow('Border', image)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
# to improve the train speed of the model 
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\Saved Pictures\wallpaperflare.com_wallpaper.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
b, g, r = cv2.split(image_rgb)

b_normalized = cv2.normalize(b.astype('float'), None, 0, 1, cv2.NORM_MINMAX)
g_normalized = cv2.normalize(g.astype('float'), None, 0, 1, cv2.NORM_MINMAX)
r_normalized = cv2.normalize(r.astype('float'), None, 0, 1, cv2.NORM_MINMAX)

normalized_image = cv2.merge((b_normalized, g_normalized, r_normalized))
print(normalized_image[:, :, 1])

normalized_image_uint8 = (normalized_image * 255).astype('uint8')
val = cv2.resize(normalized_image_uint8, (1050, 750))
cv2.imshow("Normalized Image", val)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
